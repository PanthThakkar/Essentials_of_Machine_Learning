```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import math
```

```python
def f(my_x, my_y):
    return my_x ** 2 - my_y ** 2
```

```python
xs = np.linspace(-3, 3, 1000)
```

```python
zs_wrt_x = f(xs, 0)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgrey')
plt.axhline(y=0, color='lightgrey')
_ = ax.plot(xs, zs_wrt_x)
```

```python
ys = np.linspace(-3, 3, 1000)
```

```python
as_wrt_y = f(0, ys)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgrey')
plt.axhline(y=0, color='lightgrey')
_ = ax.plot(ys, as_wrt_y)
```

```python
zs_wrt_y = f(2, ys)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgrey')
plt.axhline(y=0, color='lightgrey')
_ = ax.plot(ys, zs_wrt_y)
```

```python
def delz_dely(my_x, my_y):
    return -2 * my_y
```

```python
y_samples = [-2, -1, 0, 1, 2]
```

```python
def point_and_tangent_wrt_y(my_ys, my_x, my_y, my_f, fprime, col):
    
    my_z = my_f(my_x, my_y)
    plt.scatter(my_y, my_z, c=col)

    tangent_m = fprime(my_x, my_y)
    tangent_b = my_z - tangent_m * my_y
    tangent_line = tangent_m *my_ys + tangent_b

    plt.plot(my_ys, tangent_line, c=col, linestyle='dashed', linewidth=0.7)
```

```python
colors = ['red', 'orange', 'green', 'blue', 'purple']
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgrey')
plt.axhline(y=0, color='lightgrey')

for i, y in enumerate(y_samples):
    point_and_tangent_wrt_y(ys, 2, y, f, delz_dely, colors[i])
plt.ylim(-5, 5)

_ = ax.plot(xs, zs_wrt_y)
```

Partial Derivatives of a Cylinder's Volume - can be described by v = pi * r ^ 2 * l where r is the radius of the cylinder and L is the length of the cylinder

```python
def cylinder_vol(my_r, my_l):
    return math.pi * my_r ** 2 * my_l
```

```python
r = torch.tensor(3.).requires_grad_()
r
```

```python
l = torch.tensor(5.).requires_grad_()
l
```

```python
v = cylinder_vol(r, l)
v
```

```python
v.backward()
```

```python
l.grad
```
