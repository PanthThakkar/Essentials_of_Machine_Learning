Delta Method

```python
def diff_demo(my_f, my_x, my_delta):
    return (my_f(my_x + my_delta) - my_f(my_x)) / my_delta
```

```python
deltas = [1, 0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001]
```

```python
for delta in deltas:
    print(diff_demo(f, 2, delta))
```

```python
for delta in deltas:
    print(diff_demo(f, -1, delta))
```

Automatic Differentiation - is an application of the chain rul (typically for partial derivatives) to sequence (forward pass) of arithmetic operations

```python
import torch
```

```python
x = torch.tensor(5.0)
x
```

```python
x.requires_grad_()
```

```python
y = x ** 2
y
```

```python
y.backward()
```

```python
x.grad
```

Autodiff with TensorFlow

```python
import tensorflow as tf
```

```python
x = tf.Variable(5.0)
x
```

```python
with tf.GradientTape() as t:
    t.watch(x)
    y = x ** 2
```

```python
t.gradient(y, x)
```
